```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np


# NOTE: This module is provided by the 'python-louvain' package.
# Run: pip install python-louvain if community detection is required.
try:
    import community.community_louvain as community_louvain
except ImportError:
    print("WARNING: 'python-louvain' not found. Community detection will be skipped.")
    community_louvain = None
```

```python
# Ensure these files are in the same directory as the script
nodes_df = pd.read_csv("trump-nodes.csv")
edges_df = pd.read_csv("trump-edges.csv")
```

```python
# Build the NetworkX Graph (Undirected, including relationship attribute)
G = nx.from_pandas_edgelist(
    edges_df,
    source='Source',
    target='Target',
    edge_attr='Relationship',
    create_using=nx.Graph()
)
print(f"Graph G created with {G.number_of_nodes()} nodes and {G.number_of_edges()} edges.")
```

```python
# Calculate ALL Centrality Measures
degree_centrality = nx.degree_centrality(G)
betweenness_centrality = nx.betweenness_centrality(G)
closeness_centrality = nx.closeness_centrality(G)
try:
    # Eigenvector centrality can fail on certain graphs; handle error gracefully
    eigenvector_centrality = nx.eigenvector_centrality(G, max_iter=5000)
except nx.PowerIterationFailedConvergence:
    eigenvector_centrality = {node: 0 for node in G.nodes()}
local_clustering = nx.clustering(G)
node_degree = dict(G.degree()) # Actual number of connections
```

```python
# --- 2. Structural Analysis: Bridges and Communities (Louvain) ---

# Bridges
bridges = list(nx.bridges(G))
num_bridges = len(bridges)
print(f"\nTotal Number of Bridges: {num_bridges}")

# Communities (Louvain)
partition = {}
num_communities = 0
if community_louvain:
    partition = community_louvain.best_partition(G, weight='weight')
    num_communities = len(set(partition.values()))
    print(f"Number of Network Communities (Louvain): {num_communities}")
```

```python
# --- Calculate Degree Assortativity ---
assortativity = nx.degree_assortativity_coefficient(G)

print("\n--- Degree Assortativity Analysis ---")
print(f"Degree Assortativity Coefficient: {assortativity:.3f}")

if assortativity > 0:
    print("Interpretation: The graph is **Assortative** (nodes with high degree tend to connect to other nodes with high degree). This is common in social networks (rich-club phenomenon).")
elif assortativity < 0:
    print("Interpretation: The graph is **Disassortative** (nodes with high degree tend to connect to nodes with low degree). This is typical of a central-hub or star-like network (e.g., Donald Trump connecting to many low-degree individuals).")
else:
    print("Interpretation: The graph is neutral (connections are random with respect to degree).")
```

```python
# --- 3. Integrate Metrics, Tabulate Top Clustering, and Save Full CSV ---

metrics_df = pd.DataFrame({
    'Id': list(G.nodes()),
    'Degree Centrality': [degree_centrality.get(node, 0) for node in G.nodes()],
    'Betweenness Centrality': [betweenness_centrality.get(node, 0) for node in G.nodes()],
    'Closeness Centrality': [closeness_centrality.get(node, 0) for node in G.nodes()],
    'Eigenvector Centrality': [eigenvector_centrality.get(node, 0) for node in G.nodes()],
    'Local Clustering Coefficient': [local_clustering.get(node, 0) for node in G.nodes()],
    'Number of Connections (Degree)': [node_degree.get(node, 0) for node in G.nodes()],
    'Community_Id': [partition.get(node, -1) for node in G.nodes()]
})
metrics_df = pd.merge(nodes_df, metrics_df, on='Id', how='right')
metrics_df_sorted = metrics_df.sort_values(by='Degree Centrality', ascending=False).reset_index(drop=True)
```

```python
# Save the full metrics DataFrame to CSV
metrics_df.to_csv("trump_node_metrics.csv", index=False)
print("\nSaved all node metrics (including Community ID) to trump_node_metrics.csv")
```

```python
# Display Local Clustering Coefficient for Top 10 Nodes
top_10_clustering = metrics_df_sorted.head(10)[['Label', 'Local Clustering Coefficient']]
print("\nLocal Clustering Coefficient for Top 10 Nodes (by Degree Centrality):\n")
print(top_10_clustering.to_markdown(index=False))
```

```python
# --- 4. Final Tabulation: Top 5 Most Popular Nodes ---

# Select the Top 5 Nodes
top_5_nodes = metrics_df_sorted.head(5).copy()

final_tabulation = top_5_nodes[[
    'Label',
    'Number of Connections (Degree)', # THIS IS THE CRITICAL COLUMN
    'Degree Centrality',
    'Eigenvector Centrality',
    'Betweenness Centrality'
]].copy()

final_tabulation.columns = [
    'Node',
    'Connections (Degree)',
    'Degree Centrality',
    'Eigenvector Centrality',
    'Betweenness Centrality'
]

metric_cols = ['Degree Centrality', 'Eigenvector Centrality', 'Betweenness Centrality']
for col in metric_cols:
    final_tabulation[col] = final_tabulation[col].apply(lambda x: f'{x:.4f}')

print("\n--- Final Tabulation: Top 5 Most Popular Nodes ---\n")
print(final_tabulation.to_markdown(index=False))

final_tabulation.to_csv("top_5_popularity_tabulation.csv", index=False)
print("\nSaved final tabulation to top_5_popularity_tabulation.csv")
```

```python
# --- 5. Generate HISTOGRAMS for SPECIFIC Centrality Metrics ---

metrics_to_plot = [
    'Degree Centrality', 'Closeness Centrality', 'Betweenness Centrality',
    'Eigenvector Centrality'
]

print("\nGenerating histograms for the requested 4 centrality measures...")

for metric in metrics_to_plot:
    file_name = f'{metric.lower().replace(" ", "_")}_histogram_final.png'
    
    data = metrics_df[metric].values
    non_zero_data = data[data > 0] 
    bins = 15

    plt.figure(figsize=(8, 5))
    
    if len(non_zero_data) > 1:
        plt.hist(non_zero_data, bins=bins, color='#469990', edgecolor='black', alpha=0.7)
        plt.xlabel(metric, fontsize=12)
        plt.ylabel('Frequency (Number of Nodes)', fontsize=12)
        plt.title(f'Distribution of {metric} (Excluding Zero Scores)', fontsize=14)
        
        max_score = np.max(non_zero_data)
        plt.text(0.95, 0.95, f'Max Score: {max_score:.4f}', 
                 transform=plt.gca().transAxes, fontsize=10, 
                 verticalalignment='top', horizontalalignment='right')
    else:
        plt.text(0.5, 0.5, f'Insufficient data for histogram after excluding zero scores.', 
                 transform=plt.gca().transAxes, fontsize=12, horizontalalignment='center')
        
    plt.tight_layout()
    plt.savefig(file_name)
    plt.show() 
    plt.close()
    print(f"Saved histogram: {file_name}")
```

```python
# --- 6. Plot Full Network Graph (NO Edge Labels - Too Cluttered) ---

plt.figure(figsize=(18, 18))
pos = nx.kamada_kawai_layout(G)
node_size = [v * 10000 for v in degree_centrality.values()]

nx.draw_networkx_nodes(G, pos, node_color='skyblue', node_size=node_size, alpha=0.8)
nx.draw_networkx_edges(G, pos, edge_color='gray', alpha=0.3)

top_5_nodes_id = metrics_df_sorted.head(5)['Id'].tolist()
top_5_labels = {node: node for node in top_5_nodes_id}

nx.draw_networkx_labels(G, pos, labels=top_5_labels, font_size=10, font_color='black', font_weight='bold')

plt.title("TrumpWorld Social Network Graph (Node Size Proportional to Degree Centrality)", fontsize=18)
plt.axis('off')
plt.tight_layout()
plt.savefig("trump_network_graph.png")
plt.show() 
plt.close()
print("Saved full network graph visualization to trump_network_graph.png")

```

```python
# --- 7. Plot Ego Network Graph for TOP 3 Nodes (INCLUDES Shortened Edge Labels - Relationship) ---

top_3_nodes = metrics_df_sorted.head(3)['Label'].tolist()

print("\nGenerating Ego Network plots for the Top 3 Nodes (Includes Shortened 'Relationship' Edge Labels)...")

# ... (Visualization code here, as provided previously) ...
for EGO_NODE in top_3_nodes:
    ego_nodes = [EGO_NODE] + list(G.neighbors(EGO_NODE))
    Ego_G = G.subgraph(ego_nodes).copy()

    ego_degree_centrality = nx.degree_centrality(Ego_G)
    
    max_degree = max(ego_degree_centrality.values()) if ego_degree_centrality else 0
    if max_degree == 0:
        ego_node_size = [1000 for node in Ego_G.nodes()]
    else:
        ego_node_size = [v * 5000 / max_degree if max_degree > 0 else 1000 for v in ego_degree_centrality.values()]
        
    ego_labels = {node: node for node in Ego_G.nodes()}

    plt.figure(figsize=(12, 12))
    pos = nx.spring_layout(Ego_G, k=0.3, iterations=50)

    node_colors = ['#e6194b' if node == EGO_NODE else 'lightcoral' for node in Ego_G.nodes()]
    
    nx.draw_networkx_nodes(Ego_G, pos, node_color=node_colors, node_size=ego_node_size, alpha=0.9)
    nx.draw_networkx_edges(Ego_G, pos, edge_color='gray', alpha=0.7, width=1.5)
    nx.draw_networkx_labels(Ego_G, pos, labels=ego_labels, font_size=8, font_color='black', font_weight='bold')

    # Edge Labeling is applied here
    # 1. Get the original edge attributes
    original_edge_labels = nx.get_edge_attributes(Ego_G, 'Relationship')
    
    # 2. Shorten the labels to a maximum of 10 characters
    # If the label is longer than 10 chars, it's truncated and '...' is added.
    MAX_LENGTH = 10
    shortened_edge_labels = {
        (u, v): f"{label[:MAX_LENGTH]}..." if len(label) > MAX_LENGTH else label
        for (u, v), label in original_edge_labels.items()
    }

    nx.draw_networkx_edge_labels(
        Ego_G,
        pos,
        edge_labels=shortened_edge_labels, # Use the shortened labels here
        font_size=7,
        label_pos=0.3, 
        bbox={"boxstyle": "round,pad=0.2", "fc": "white", "alpha": 0.5}
    )

    file_name = f"{EGO_NODE.lower().replace(' ', '_').replace('.', '')}_ego_network_final.png"
    plt.title(f"Ego Network of {EGO_NODE}", fontsize=16)
    plt.axis('off')
    plt.tight_layout()
    plt.savefig(file_name)
    plt.show()
    plt.close()
    print(f"Saved ego network plot for {EGO_NODE} to {file_name}")
```

```python
# --- 8. Plot Community Network (Only runs if python-louvain is available) ---

if community_louvain:
    print("\nGenerating Community Network Visualization...")
    
    pos = nx.spring_layout(G, k=0.15, iterations=50)
    community_map = [partition[node] for node in G.nodes()]
    
    plt.figure(figsize=(20, 20))
    node_size = [v * 10000 for v in degree_centrality.values()]
    
    cmap = plt.cm.get_cmap('Spectral', num_communities)
    
    nx.draw_networkx_nodes(
        G, pos, 
        node_color=community_map, 
        node_size=node_size, 
        cmap=cmap, 
        alpha=0.8
    )
    nx.draw_networkx_edges(G, pos, alpha=0.1)

    top_5_nodes_comm = sorted(degree_centrality, key=degree_centrality.get, reverse=True)[:5]
    top_5_labels_comm = {node: node for node in top_5_nodes_comm}
    nx.draw_networkx_labels(G, pos, labels=top_5_labels_comm, font_size=10, font_weight='bold', font_color='black')
    
    modularity_score = community_louvain.modularity(partition, G, weight='weight')
    plt.title(f"Network Communities (Louvain, Modularity: {modularity_score:.3f})", fontsize=20)
    plt.axis('off')
    plt.tight_layout()
    plt.savefig("community_network_plot.png")
    plt.show()
    plt.close()
    print("Saved Community Network Visualization to community_network_plot.png")
```

```python
import networkx as nx
import pandas as pd
import community.community_louvain as community_louvain
from networkx.algorithms.community.centrality import girvan_newman
import itertools
import numpy as np

# --- Import igraph for Leiden and SBM ---
try:
    import igraph as ig
except ImportError:
    print("WARNING: 'python-igraph' not found. Leiden and SBM will be skipped.")
    ig = None

# [cite_start]--- ASSUMED BASELINE VALUES from your PDF [cite: 47, 441] ---
# G (Your NetworkX graph, 303 nodes, 366 edges)
modularity_score_standard = 0.614
num_communities_standard = 32

comparison_results = {
    'Louvain (Standard)': (num_communities_standard, modularity_score_standard)
}
# ------------------------------------------------------------------

# 1. Louvain with Resolution Parameter (gamma = 1.5)
print("\n--- 1. Louvain (Resolution=1.5) Results ---")

resolution_parameter = 1.5
# Find partition using custom resolution
partition_res = community_louvain.best_partition(G, weight='weight', resolution=resolution_parameter)
# Fix 1: modularity() function does not accept 'resolution' in this package version.
modularity_res = community_louvain.modularity(partition_res, G, weight='weight')
num_communities_res = len(set(partition_res.values()))

print(f"Number of Communities: {num_communities_res}")
print(f"Modularity Score (of Res. 1.5 Partition): {modularity_res:.4f}")
comparison_results['Louvain (Res. 1.5)'] = (num_communities_res, modularity_res)


# 2. Girvan-Newman Algorithm (Highest Modularity Partition)
print("\n--- 2. Girvan-Newman Algorithm ---")

gn_iterator = girvan_newman(G)
max_gn_modularity = -1.0
best_gn_communities = None
target_divisions = 10 # Limit iterations for efficiency

for i, communities in enumerate(gn_iterator):
    if i >= target_divisions:
        break
    
    current_modularity = nx.community.modularity(G, communities, weight='weight')
    
    if current_modularity > max_gn_modularity:
        max_gn_modularity = current_modularity
        best_gn_communities = communities

num_gn_communities = len(best_gn_communities) if best_gn_communities else 0

print(f"Number of Communities (Best of first {target_divisions} splits): {num_gn_communities}")
print(f"Max Modularity Score found: {max_gn_modularity:.4f}")
comparison_results['Girvan-Newman'] = (num_gn_communities, max_gn_modularity)


# 3. Leiden 
if ig:
    # Convert NetworkX graph to iGraph format
    edges = list(G.edges(data=True))
    edge_list = [(u, v) for u, v, data in edges]
    edge_weights = [data.get('weight', 1.0) for u, v, data in edges]

    ig_graph = ig.Graph.TupleList(edge_list, directed=False)
    ig_graph.es["weight"] = edge_weights
    node_names = list(G.nodes())

    # 3. Leiden Algorithm
    print("\n--- 3. Leiden Algorithm (Robust & Fast) ---")
    leiden_partition_ig = ig_graph.community_leiden(
        objective_function='modularity', 
        weights='weight'
    )
    leiden_modularity = leiden_partition_ig.modularity
    num_leiden_communities = len(leiden_partition_ig)
    
    print(f"Number of Communities: {num_leiden_communities}")
    print(f"Modularity Score: {leiden_modularity:.4f}")
    comparison_results['Leiden'] = (num_leiden_communities, leiden_modularity)
```

```python
# --- 9. Prepare Data for Gephi (GEXF) ---

for node, data in G.nodes(data=True):
    data['Degree_Centrality'] = degree_centrality.get(node, 0)
    data['Betweenness_Centrality'] = betweenness_centrality.get(node, 0)
    data['Closeness_Centrality'] = closeness_centrality.get(node, 0)
    data['Eigenvector_Centrality'] = eigenvector_centrality.get(node, 0)
    data['Local_Clustering'] = local_clustering.get(node, 0)
    data['Community_Id'] = partition.get(node, -1) 

    node_label = nodes_df[nodes_df['Id'] == node]['Label'].iloc[0] if node in nodes_df['Id'].values else node
    data['Label'] = node_label

nx.write_gexf(G, "trump_network_gephi.gexf")
print("\nSaved graph for Gephi import to trump_network_gephi.gexf")
```
